```python
from sklearn.manifold import TSNE
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
import matplotlib.pyplot as plt
import umap.umap_ as umap
```

```python
df = pd.read_csv('haberman.data', sep=',', usecols = ["Age","Patient's year of operation","Number of positive axillary nodes detected","Survival status"])
df
```

```python
cl = df["Survival status"]
df1 = df[["Age","Patient's year of operation","Number of positive axillary nodes detected"]]
df1
```

```python
x_axis = np.array(range(df.shape[0]))
a = np.array(df['Age'])
b = np.array(df["Patient's year of operation"])
c = np.array(df['Number of positive axillary nodes detected'])
```

```python
plt.plot(x_axis, a, color='r', label='Age')
plt.plot(x_axis, b, color='g', label='Patients year of operation')
plt.plot(x_axis, c, color='b', label='Number of positive axillary nodes detected')
plt.xlabel('instances')
plt.ylabel('value')
plt.legend()
plt.show()
```

```python
def tsne(dataset):
    X_embedded = TSNE(n_components=2, learning_rate='auto', init='random', perplexity=3).fit_transform(dataset)
    return X_embedded
```

```python
def build_graph(axis,value):
    col = []
    x = []
    y = []
    for i in range(0,len(value)):
        if value[i] == 1:
            col.append('green')
        elif value[i] == 2:
            col.append('red')
    for i in range(len(axis)):
        x.append(axis[i][0])
        y.append(axis[i][1])
    for i in range(len(x)):
        plt.scatter(x[i], y[i], c = col[i], s = 10, linewidth = 0)
    plt.show()
```

```python
value = df['Survival status'].tolist()
```

```python
axis = tsne(df1.to_numpy())
```

```python
build_graph(axis,value)
```

```python
axis = tsne(x_minmax)
```

```python
build_graph(axis,value)
```

```python
axis = tsne(x_standard)
```

```python
build_graph(axis, value)
```

```python
axis = tsne(x_robust)
```

```python
build_graph(axis, value)
```
